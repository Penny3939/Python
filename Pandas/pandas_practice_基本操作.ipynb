# 時間處理 

```python
#秒數=>格林威治時區

from datetime import datetime
str(datetime.utcfromtimestamp(1140825600))  #轉換成字串
```

```python
#要對整份文件修改
#轉換流程 -> 遞給整行 ->每一個走過流程
#利用 .apply(流程名字)對每一個執行一次流程

def timetrans(s):
    return str(datetime.utcfromtimestamp(s)) 

df["film_utc"] = df['film_date'].apply(timetrans) #將值傳至film_utc
df
```

```python
def fflow(n):
    if n>3000 and n<5000:
        return True
    else:
        return False

f= df["comments"].apply(fflow)
df[f]
```

```python
df ["tags"]
```

```python
#eval() => 把括號裡的字串當成一個python程式執行
# eval("print(3+2)") ->印出5
# a = eval("[1,2,3]") ->a=[1,2,3]


def tagtrans(t):
    taglist= eval(t)
    if "children" in taglist:
        return True
    else:
        return False
df[df["tags"].apply(tagtrans)]
```
