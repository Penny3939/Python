# Missing Data

```python
import numpy as np
import pandas as pd

d= {"A":[1,2,np.nan], "B":[5,np.nan, np.nan], "C":[1,2,3]}
d
```

```python
df = pd.DataFrame(d)
df
```

```python
#fillna 取代nan空值
df.fillna(value="abc")
```

```python
#練習一:
#指定A欄位的空值用"Hello"取代

df["A"].fillna(value="Hello") 
```

```python
#練習二:
#指定A、B欄位的空值用"Hello"取代

df[ ["A","B"] ].fillna(value="Hello")
```

```python
#練習三:
#A欄位裡的空值,用A欄位的平均值取代

df["A"].fillna(value= df["A"].mean())
```

```python
df
```

```python
#刪除空值 drop

df.dropna()  #預設axis=0
```

```python
df.dropna(axis=1)
```

```python
#thresh: 只要有兩個有效數值,就print出結果

df.dropna(thresh=2)  #預設axis=0
```

```python
#thresh: 只要有兩個有效數值,就print出結果

df.dropna(thresh=2, axis=1) 
```

# Group by

```python
import numpy as np
import pandas as pd

data= {"Company":["Google", "Google", "Ig", "Ig","Fb","Fb"], 
       "Person":["Sam","Charlie","Amy", "Penny","Carlo","Sarah"], 
       "Sales":[200,120,340,124,243,350]}
df = pd.DataFrame(data)
df
```

```python
#練習一: 依公司名加總

by_company = df.groupby("Company")
by_company.sum()
```

```python
#依公司名加總,並指定company= "FB"

# 方法一:
df.groupby("Company").sum().loc["Fb"]
    

# 方法二:
# by_company = df.groupby("Company")
# by_company.sum().loc["Fb"]
```

```python
#練習二: 依公司名算平均值

by_company = df.groupby("Company")
by_company.mean()
```

```python
#練習二: 依公司名算平均值, 並指定company=Google

#方法一:
df.groupby("Company").mean().loc["Google"]


#方法二:
# by_company = df.groupby("Company")
# by_company.mean().loc["Google"]
```

```python
#練習三 : 計算欄位有幾個值

df.groupby("Company").count()
```

```python
df
```

```python
#練習四:max/ min

df.groupby("Company").min()
```

```python
df.groupby("Company").max()
```

```python
# 依公司名稱列出以下
df.groupby("Company").describe()
```

```python
#指定公司為Fb

# 方法一:
df.groupby("Company").describe().loc["Fb"]

# 方法二: 不使用loc
# df.groupby("Company").describe().transpose()["Fb"]
```

# concat 連接

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame( {"A":["a0","a1","a2","a3"], 
                     "B":["b0","b1","b2","b3"], 
                     "C":["c0","c1","c2","c3"],
                     "D":["d0","d1","d2","d3"]}, index = [0,1,2,3])

df2 = pd.DataFrame( {"A":["a4","a5","a6","a7"], 
                     "B":["b4","b5","b6","b7"], 
                     "C":["c4","c5","c6","c7"],
                     "D":["d4","d5","d6","d7"]}, index = [4,5,6,7])

df3 = pd.DataFrame( {"A":["a8","a9","a10","a11"], 
                     "B":["b8","b9","b10","b11"], 
                     "C":["c8","c9","c10","c11"],
                     "D":["d8","d9","d10","d11"]}, index = [8,9,10,11])
```

```python
pd.concat( [df1, df2, df3] )
```

```python
pd.concat( [df1, df2, df3], axis=1)
```

# Merging (合併)

```python
import numpy as np
import pandas as pd

left = pd.DataFrame({"key": ["k0","k1","k2","k3"],
                    "A":["a0","a1","a2","a3"],
                    "B":["b0","b1","b2","b3"]})

right = pd.DataFrame({"key": ["k0","k1","k2","k3"],
                    "C":["c0","c1","c2","c3"],
                    "D":["d0","d1","d2","d3"]})
```

```python
left
```

```python
right
```

```python
# merge合併(依columns)
pd.merge(left,right,how= "inner", on="key")  #inner表示交集部分

# pd.merge(left,right,how= "outer", on="key")  #outer表示列出所有key
# pd.merge(left,right,how= "right", on= "key") #right表示列出所有right key
```

# Join

```python
import numpy as np
import pandas as pd

left = pd.DataFrame({"A":["a0","a1","a2"],
                    "B":["b0","b1","b2"]}, index = ["k0","k1","k2"])

right = pd.DataFrame({"C":["c0","c1","c2"],
                    "D":["d0","d1","d2"]}, index = ["k0","k2","k3"])
```

```python
left
```

```python
right
```

```python
#依left表為主

left.join(right)
```

```python
#依right表為主

right.join(left)
```

```python
#列出所有資料how="outer"

left.join(right, how="outer")
```
